# Unstructured data:

* import numpy to deal with Null values:

---



```python
import numpy as np
```

```python
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
```

```python
data
```

* convert into Structured data using DataFrame:

---



```python
x = pd.DataFrame(data)
```

```python
x
```

* Operations:

---



```python
x['visits']>2
```

```python
x[x['visits']>2]
```

```python
x[x['animal'] == 'snake']
```

* '&' operator:

---



```python
x[ (x['animal'] == 'cat') & (x['visits'] <3) ]
```

* ranging for column values:

---



```python
x[x['visits'].between(1,5)]
```
